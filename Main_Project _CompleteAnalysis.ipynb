# <span style="color:green">**6. HANDLING MISSING DATA**</span>

#### 1. Check for missing datas

```python
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)
```

#### 2. Missing data in the column "Primary Customer City" to be replaced by the most repeated value of the column (mode)

```python
mode_city=df['Primary Customer City'].mode()[0]
mode_city
```

STAMFORD is the most repeated city. So filling the missing value in the column with STAMFORD.

```python
df.fillna({"Primary Customer City" : mode_city}, inplace=True)
```

#### 3. Missing data in the column "Primary Customer State" to be replaced by the most repeated value of the column (mode)

```python
mode_state=df['Primary Customer State'].mode()[0]
mode_state
```

CT is the most repeated state. So filling the missing value in the column with CT

```python
df.fillna({"Primary Customer State" : mode_state}, inplace=True)
```

#### 4. Missing data in the column "Vehicle Category" to be replaced by the most repeated value of the column (mode)

```python
mode_category=df['Vehicle Category'].mode()[0]
mode_category
```

Light-Duty (Class 1-2) is the most repeated category. So filling the missing value in the column with 'Light-Duty (Class 1-2)'.

```python
df.fillna({"Vehicle Category" : mode_category}, inplace=True)
```

#### 5. Handling datetime columns

The columns "Registration start date" and "Registration Expiration date" are datetime columns. Since the dates in the dataset are in series it will be logical to handle it with forward fill (ffill ) method. Here Iam using ffill() method. The ffill() method replaces the NULL values with the value from the previous row.

```python
df[['Registration Start Date', 'Registration Expiration Date']] = \
    df[['Registration Start Date', 'Registration Expiration Date']].ffill()
```

```python
df.isnull().sum()
```
